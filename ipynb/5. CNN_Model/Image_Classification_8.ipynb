#### Custom train_test split

```python
def custom_train_test_split(dataset, train_ratio, val_ratio, test_ratio):
  ds_len = len(dataset)
  sub_train = dataset.take(int(ds_len*train_ratio))
  sub_val_test = dataset.skip(int(ds_len*train_ratio))
  sub_val = sub_val_test.take(int(ds_len*val_ratio))
  sub_test = sub_val_test.skip(int(ds_len*val_ratio))
  return sub_train, sub_val, sub_test
```

##### Using it on train_data_gen dataset
- For testing purpose only

```python
dataset = train_data_gen
TRAIN_RATIO = 0.6
VAL_RATIO = 0.2
TEST_RATIO = 0.2
data_train, data_val, data_test = custom_train_test_split(dataset, TRAIN_RATIO, VAL_RATIO, TEST_RATIO)
print(data_train)
print(data_val)
print(data_test)
print(list(data_train.take(1).as_numpy_iterator()))
print(list(data_val.take(1).as_numpy_iterator()))
print(list(data_test.take(1).as_numpy_iterator()))
```

## 7. Building First CNN model

#### Import relevant libraries

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, InputLayer, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.metrics import Accuracy, Precision, Recall, AUC
```

#### Model Initializing

```python
first_model = Sequential(name="Plain_Model")

first_model.add(InputLayer(input_shape=(IMG_SIZE, IMG_SIZE, 3), name="Input_Layer"))

first_model.add(Conv2D(filters=32, kernel_size=3, activation="relu", padding="same", name="First_Conv2D_32"))
first_model.add(MaxPool2D(pool_size=2, name="First_MaxPool2D_2"))

first_model.add(Conv2D(filters=64, kernel_size=3, activation="relu", padding="same", name="Second_Conv2D_32"))
first_model.add(MaxPool2D(pool_size=2, name="Second_MaxPool2D_2"))

first_model.add(Conv2D(filters=128, kernel_size=3, activation="relu", padding="same", name="Third_Conv2D_32"))
first_model.add(MaxPool2D(pool_size=2, name="Third_MaxPool2D_2"))

first_model.add(Flatten())

first_model.add(Dense(256, activation="relu", name="First_Dense_Layer_256"))
first_model.add(Dense(128, activation="relu", name="Second_Dense_Layer_128"))

first_model.add(Dense(9, activation="softmax", name="Final_Layer_9"))
```

#### Model Summary

```python
first_model.summary()
```

#### Model Compiling

```python
first_model.compile(optimizer=Adam(learning_rate=0.001), loss=CategoricalCrossentropy(),
                    metrics=[Accuracy(), Precision(), Recall(), AUC()])
```

#### Model Fitting

```python
%%time
first_model_history = first_model.fit(train_data_gen_prefetched,
                                      epochs=30,
                                      validation_data=test_data_gen_prefetched)
```

## 8. Building Second CNN model with Dropout layers

#### Model Initializing

```python
second_model = Sequential(name="Plain_Model_2")

second_model.add(InputLayer(input_shape=(IMG_SIZE, IMG_SIZE, 3), name="Input_Layer"))

second_model.add(Conv2D(filters=32, kernel_size=3, activation="relu", padding="same", name="First_Conv2D_32"))
second_model.add(MaxPool2D(pool_size=2, name="First_MaxPool2D_2"))
second_model.add(Dropout(0.2, name="CNN_First_Dropout_0.2"))

second_model.add(Conv2D(filters=64, kernel_size=3, activation="relu", padding="same", name="Second_Conv2D_32"))
second_model.add(MaxPool2D(pool_size=2, name="Second_MaxPool2D_2"))
second_model.add(Dropout(0.2, name="CNN_Second_Dropout_0.2"))

second_model.add(Conv2D(filters=128, kernel_size=3, activation="relu", padding="same", name="Third_Conv2D_32"))
second_model.add(MaxPool2D(pool_size=2, name="Third_MaxPool2D_2"))
second_model.add(Dropout(0.2, name="CNN_Third_Dropout_0.2"))

second_model.add(Flatten())

second_model.add(Dense(128, activation="relu", name="First_Dense_Layer_128"))
second_model.add(Dropout(0.4, name="ANN_First_Dropout_0.4"))
second_model.add(Dense(256, activation="relu", name="Second_Dense_Layer_256"))
second_model.add(Dropout(0.4, name="ANN_Second_Dropout_0.4"))
second_model.add(Dense(128, activation="relu", name="Third_Dense_Layer_128"))
second_model.add(Dropout(0.4, name="ANN_Third_Dropout_0.4"))

second_model.add(Dense(9, activation="softmax", name="Final_Layer_9"))
```

#### Model Summary

```python
second_model.summary()
```

#### Model Compiling

```python
second_model.compile(optimizer=Adam(learning_rate=0.001), loss=CategoricalCrossentropy(),
                    metrics=[Accuracy(), Precision(), Recall(), AUC()])
```

#### Model Fitting

```python
%%time
second_model_history = second_model.fit(train_data_gen_prefetched,
                                      epochs=30,
                                      validation_data=test_data_gen_prefetched)
```
